# Construtor
***

**list()**: Pega qualquer tipo de dado iteravel e transforma em uma lista

***
#### Exemplos
***

```python
# Construtor com listas
print(list([1,2,3,4]))
```

***

```python
# Construtor com texto 
print(list("auhauhauah"))
```

***

```python
# Construtor com range
print(list(range(1,5)))
```
